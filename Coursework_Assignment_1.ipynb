Exercise 3
Plot the function y = x^2 over the domain [0, 10) in increments of
0.2
.

```python
x_axis = np.arange(0, 10, 0.2)
y_axis = [v*v for v in x_axis]
x_axis, y_axis
#plt.figure(figsize=(20, 8))
plt.plot( x_axis, y_axis )
plt.title("function: y=x^2")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

Exercise 14
Read the documentation of the numpy.where() function. Use numpy.where()Â function to create a new sequence, substituting -1 in place of 0 in the result from the previous exercise.

```python
binary_seq = np.where(Bernoulli_seq==0, -1, Bernoulli_seq)
binary_seq
```

Exercise 15

```python
cumulative_seq = np.cumsum(binary_seq)
#plt.figure(figsize=(20, 8))
plt.plot( cumulative_seq )
plt.title(r'$y=\sum_{j=1}^i \epsilon_j$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

Exercise 16
Plot n=20 independent realisations of the above process on the same graph in different colours. Your solution should not make use of a for loop.


```python
Bernoulli_seq = np.random.randint(2, size=100*20)
seqs = Bernoulli_seq.reshape(-1, 20)
seqs = np.cumsum(seqs, axis=0)
import pandas as pd
Bernoulli_df = pd.DataFrame(seqs, index=list(range(0,100)))
Bernoulli_df
Bernoulli_df.plot.line()
```

```python
rs = np.random.normal(0, 0.005, 100*20)
rs = rs.reshape(-1, 20)
seqs = np.cumsum(rs, axis=0)
Y = np.exp(seqs)
import pandas as pd
stochastic_df = pd.DataFrame(Y, index=list(range(0,100)))
stochastic_df
stochastic_df.plot.line()
```
